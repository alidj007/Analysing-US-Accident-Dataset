##  Exploratory Data Analysis on column 'Start Time'

```python
df.Start_Time
```

```python
df.Start_Time[0]
```

```python
df.Start_Time.dtype
```

```python
## Converting String date to Timestamp format
df.Start_Time = pd.to_datetime(df.Start_Time)
```

```python
df.Start_Time.dtype
```

```python
df.Start_Time[0]
```

```python
## Extracting Hours from Date_Time column
df.Start_Time.dt.hour
```

```python
## Plotting histogram of distribution of accidents hour wise
sns.histplot(df.Start_Time.dt.hour,bins=24);
```

```python
## Plotting distribution of hourly accident percentage wise
sns.distplot(df.Start_Time.dt.hour,bins=24, kde=False, norm_hist=True);
```

    ## Insight from Graph - A high number of accidents occur between 6am to 10am and 3pm to 6pm(probably during office hour)

```python
## Plotting distribution of accidents by day 
sns.distplot(df.Start_Time.dt.dayofweek,bins=7, kde=False, norm_hist=True);
```

    ### Insight - The distribution of number of accidents is evenly distributed 

```python
## Checking whether the distribution of accidents by hour the same on weekends as on week days.
sunday = df.Start_Time[df.Start_Time.dt.dayofweek == 6]
sns.distplot(sunday.dt.hour,bins=24, kde=False, norm_hist=True);
```

    ### Insight - On Sunday the accident by hours is more evenly distributed

```python
## Plotting distribution of accidents on Mondays
monday = df.Start_Time[df.Start_Time.dt.dayofweek == 0]
sns.distplot(monday.dt.hour,bins=24, kde=False, norm_hist=True);
```

    ### Insight - Distribution of number of accidents on Mondays(week Day) follows a Bimodal pattern

```python
## Plotting distribution of accidents Month Wise.
sns.distplot(df.Start_Time.dt.month,bins=12, kde=False, norm_hist=True);
```

    ### Insight - Distribution of number of accidents month wise is more evenly distributed for first 8 months <br> 
        and then the number of accidents rises may be due to winter fogs.

```python
## Plotting distribution of accidents year wise 
year_2019 = df[df.Start_Time.dt.year == 2016]
sns.distplot(year_2019.Start_Time.dt.month, bins=12, kde=False, norm_hist=True);
```

```python
## Calculating the number of accidents in a particular month of a particular year
year_2016 = df[df.Start_Time.dt.year == 2016]
month_2016 =year_2016[year_2016.Start_Time.dt.month == 7]
month_2016
```

```python
## Printing the number of accidents in a particular month.
len(month_2016)
```
